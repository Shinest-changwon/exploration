```python
# 사진 준비하기
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os, glob
import copy

# 이미지 준비하기
img_path = os.getenv("HOME") + "/aiffel/exploration/camera_sticker/images"
img_list = glob.glob(img_path + "/*.jpg") 

img_list = sorted(img_list)
    
# OpenCV로 준비한 이미지 읽고, resize하기
img_bgr_list = list()
for img in img_list:
    tmp = cv2.imread(img)
    w = int(tmp.shape[1])
    h = int(tmp.shape[0])
    tmp_resize = cv2.resize(tmp, (w - w%10, h - h%10))
    img_bgr_list.append(tmp_resize)


# OpenCV는 이미지 채널이 BGR 순이다. RGB순서로 바꾸어야 한다.
img_rgb_list = list()
for bgr in img_bgr_list:
    img_rgb_list.append(cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB))

# 출력 이미지 만들기
img_show_list = list()
for rgb in img_rgb_list:
    img_show_list.append(copy.deepcopy(rgb))

# 이미지 출력해보기
for show in img_show_list:
    plt.imshow(show)
    plt.xticks([])   # x축 눈금 없애기
    plt.yticks([])   # y축 눈금 없애기
    plt.show()
```

```python
# 얼굴 검출 face detection

# 이미지의 단순화된 버전을 만들어 주는 HOG 알고리즘을 사용하여 사진을 인코딩합니다.
# HOG 알고리즘은 픽셀값의 변화로 파악할 수 있는 사진 밝기 변화의 방향을 그래디언트로 표현하고, 이로부터 객체의 형태를 찾아낼 수 있습니다.
# 얼굴 탐색 이외에도 보행자 검출 등에 활용할 수 있다.

import dlib   

# dlib를 활용해 hog detector를 선언
detector_hog = dlib.get_frontal_face_detector()   #- 기본 얼굴을 검출하는 detector 선언

# detector를 이용해 얼굴의 bounding box를 추출한다.
dlib_rects_dict = dict()
img_recog_list = list()
img_ignor_list = list()
i = 0
for show in img_show_list:
    if not detector_hog(show, 1):
        img_ignor_list.append(i)
        i = i + 1
    else:
        dlib_rects_dict[i] = detector_hog(show, 1)
        img_recog_list.append(i)
        i = i + 1

# 찾은 얼굴을 화면에 출력하기
# 찾은 얼굴영역 좌표 print(dlib_rects)
print("* 인식한 이미지")
for key, value in dlib_rects_dict.items():
    dlib_rects = value
    for dlib_rect in dlib_rects:
        l = dlib_rect.left()
        t = dlib_rect.top()
        r = dlib_rect.right()
        b = dlib_rect.bottom()
        
        cv2.rectangle(img_show_list[key], (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)
        
    # 이미지 출력하기
    plt.imshow(img_rgb_list[key])
    plt.xticks([])   # x축 눈금 없애기
    plt.yticks([])   # y축 눈금 없애기
    plt.show()
    
    # 이미지 출력하기
    plt.imshow(img_show_list[key])
    plt.xticks([])   # x축 눈금 없애기
    plt.yticks([])   # y축 눈금 없애기
    plt.show()

print("* 인식 못한 이미지")
for i in img_ignor_list:
    # 이미지 출력하기
    plt.imshow(img_rgb_list[i])
    plt.xticks([])   # x축 눈금 없애기
    plt.yticks([])   # y축 눈금 없애기
    plt.show()

    
```

```python
# 얼굴 랜드마크 face landmark
# face landmark localization : 이목구비의 위치를 추론하는 것
# face landmark는 detection의 결과물인 bountding box로 잘라낸 얼굴 이미지를 이용합니다.

# 지정된 landmark 모델 불러오기
model_path = os.getenv('HOME')+'/aiffel/exploration/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)

# landmark_predictor는 rgb이미지와 dlib.rectangle을 입력받아 dlib.full_object_detection을 반환합니다.
landmarks_dict = dict()
list_points_dict = dict()
for key, value in dlib_rects_dict.items():
    temp = dict()
    dlib_rects = value
    count = 0
    for dlib_rect in dlib_rects:
        if len(dlib_rects_dict[key]) == 1:
            points = landmark_predictor(img_rgb_list[key], dlib_rects_dict[key][0])
            landmarks_dict[key] = list(map(lambda p: (p.x, p.y), points.parts()))
        else:
            if count == len(dlib_rects_dict[key]) - 1:
                points = landmark_predictor(img_rgb_list[key], dlib_rects_dict[key][count])
                temp[count] = list(map(lambda p: (p.x, p.y), points.parts()))
                landmarks_dict[key] = temp
            else:
                points = landmark_predictor(img_rgb_list[key], dlib_rects_dict[key][count])
                temp[count] = list(map(lambda p: (p.x, p.y), points.parts()))
                count = count + 1

# points는 dlib.full_object_detection의 객체이기 때문에 parts() 함술 개별 위치에 접근할 수 있습니다. 
# 조금 더 직관적인 (x, y) 형태로 접근할 수 있도록 변환해 주었습니다. 따라서 list_points는 tuple(x, y) 68개로 이루어진 리스트입니다.
# 이미지에서 찾아진 얼굴 개수마다 복사하면 list_landmark에 68개의 랜드마크가 얼굴 개수만큼 저장됩니다
# 랜드마크 출력
for key, value in landmarks_dict.items():
    if len(value) == 68:
        for i in range(len(value)):
            point = value[i]
            cv2.circle(img_show_list[key], point, 2, (0, 255, 255), -1) # yellow
    else:
        for i in list(value.keys()):
            v = value[i]
            for j in range(len(v)):
                point = v[j]
                cv2.circle(img_show_list[key], point, 2, (0, 255, 255), -1) # yellow


                 
# 랜드마크 이미지 출력하기
for i in img_recog_list:
    plt.imshow(img_show_list[i])
    plt.xticks([])   # x축 눈금 없애기
    plt.yticks([])   # y축 눈금 없애기
    plt.show()
```
